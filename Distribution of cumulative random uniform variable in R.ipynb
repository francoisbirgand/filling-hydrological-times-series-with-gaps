```r
nbrun<-1000
X<-matrix(runif(nbrun*nbrun),nbrun)-0.5
Y<-apply(X,2,cumsum)
matplot(Y,type = "l")
signif(quantile(Y[dim(Y)[1],],probs = c(0.025,0.975))/dim(Y)[1]*100,3)
```

```r
nbrun=200
X<-matrix(runif(nbrun*nbrun),nbrun)-0.5
Y<-apply(X,2,cumsum)
beaminf<-signif(quantile(Y[dim(Y)[1],],probs = c(0.025,0.975))/dim(Y)[1]*100,3)[1]
beamsup<-signif(quantile(Y[dim(Y)[1],],probs = c(0.025,0.975))/dim(Y)[1]*100,3)[2]
for (nbrun in seq(1000,10000,by=1000)){
    X<-matrix(runif(nbrun*nbrun),nbrun)-0.5
    Y<-apply(X,2,cumsum)
    beaminf<-c(beaminf,signif(quantile(Y[dim(Y)[1],],probs = c(0.025,0.975))/dim(Y)[1]*100,3)[1])
    beamsup<-c(beamsup,signif(quantile(Y[dim(Y)[1],],probs = c(0.025,0.975))/dim(Y)[1]*100,3)[2])
}
par(mar=c(4.5,6,2,2))
plot(c(200,seq(1000,10000,by=1000)),beaminf,type="l",col="blue",ylim=c(-5,5), 
     xlab="number of cumulated random variable (uniform distribution)",
     ylab = "95%ile envelope of the %age error \nmade on the cumulative sum of variables")
abline(h=0)
par(new=TRUE)
plot(c(200,seq(1000,10000,by=1000)),beamsup,type="l",col="red",ylim=c(-5,5),xaxt="n",yaxt="n",xlab="",ylab="")
```
